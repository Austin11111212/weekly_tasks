```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/Africa_climate_change.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# Filter data for Tunisia and Cameroon
df_tunisia = df[df['COUNTRY'] == 'Tunisia']
df_cameroon = df[df['COUNTRY'] == 'Cameroon']
```

```python
df_tunisia
```

```python
df_cameroon
```

```python
# Plotting average temperature fluctuations
plt.figure(figsize=(10, 6))
plt.plot(df_tunisia['DATE'], df_tunisia['TAVG'], label='Tunisia')
plt.plot(df_cameroon['DATE'], df_cameroon['TAVG'], label='Cameroon')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.title('Average Temperature Fluctuations in Tunisia and Cameroon (1980-2023)')
plt.legend()
plt.grid(True)
plt.show()
```
