```python
# Attempt to infer the format automatically, handling mixed formats if necessary
df['DATE'] = pd.to_datetime(df['DATE'],  format='mixed', errors='coerce')

# Check for any parsing errors (NaT values)
print(df[df['DATE'].isnull()])
```

```python
# Filter data between 1980 and 2023
df = df[(df['DATE'].dt.year >= 1980) & (df['DATE'].dt.year <= 2023)]
```

```python
# Ensure 'TAVG' is numeric
df['TAVG'] = pd.to_numeric(df['TAVG'], errors='coerce')
```

```python
# Drop rows with missing values
df = df.dropna()
```

Selecting the Best Chart to Show Average Temperature per Country

```python
# Calculate average temperature per country
avg_temp_per_country = df.groupby('COUNTRY')['TAVG'].mean().sort_values()
```

```python
avg_temp_per_country
```

```python
fig = px.bar(avg_temp_per_country, x='COUNTRY', y='TAVG')
fig.show()
```

```python
# Plotting bar chart
plt.figure(figsize=(10, 6))
avg_temp_per_country.plot(kind='bar', color='skyblue')

plt.title('Average Temperature per Country')
plt.xlabel('COUNTRY')
plt.ylabel('Average Temperature (Â°C)')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
