```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('/content/Earthquake Perception Dataset.csv')
```

```python
fig = px.pie(df, names= 'sex', values = 'age', width = 500,
            color_discrete_sequence = px.colors.sequential.GnBu)
fig.update_traces(textposition='inside', textinfo='percent', textfont_size=20)
fig.show()
```

```python
df.head(2)
```

```python
fig = px.pie(df, names= 'marital_status', values = 'age', width = 500,
            color_discrete_sequence = px.colors.sequential.GnBu)
fig.update_traces(textposition='inside', textinfo='percent', textfont_size=20)
fig.show()
```

```python
px.histogram(df, x = 'age', width = 800, height = 500,nbins = 200, color = 'sex')
```

```python
px.scatter(df, x = 'age', y = 'fear',width = 1000,size = 'anxiety', color = 'sex')
```

```python
# you use sum to get total of a dataset
rand = df.groupby(['sex','marital_status'])['age'].mean().reset_index().sort_values(by = 'age', ascending = False)
px.funnel(rand, y = 'marital_status', x = 'age', width = 1200, color = 'sex')
```

```python
rand = df.groupby(['sex','marital_status'])['age'].mean().reset_index().sort_values(by = 'age', ascending = False)
px.bar(rand, y = 'age', x = 'marital_status', width = 1200, color = 'sex', barmode = 'group')
```
