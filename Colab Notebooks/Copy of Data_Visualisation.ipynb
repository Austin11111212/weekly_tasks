```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('/content/fake_gomycode_student_database.csv')
```

```python
df.head()
```

```python
# THE LINES IN CODE IS CALLED DATA POINT
fig = px.bar(df, x='Course', y='AmountPaid')
fig.show()
```

```python
df_course_amount = df.groupby('Course')['AmountPaid'].sum().reset_index()
df_course_amount
```

```python
fig = px.bar(df_course_amount, x='Course', y='AmountPaid')
fig.show()
```

```python
fig = px.bar(df, x='Course', y='AmountPaid', color = 'MaritalStatus')
fig.show()
```

```python
df_course_amount = df.groupby(['MaritalStatus','Course'])['AmountPaid'].sum().reset_index()
df_course_amount
```

```python
fig = px.bar(df_course_amount, x='Course', y='AmountPaid', color = 'MaritalStatus', title = 'Fake Gomycode Plot')
fig.show()
```

```python
mar_amount = df.groupby('MaritalStatus')['AmountPaid'].sum().reset_index()
mar_amount
```

```python
px.pie(mar_amount, names = 'MaritalStatus', values = 'AmountPaid', width = 400, height = 400, title = 'Fake Title')
```

```python
fig = px.pie(mar_amount, names = 'MaritalStatus', values = 'AmountPaid', width = 400, height = 400, title = 'Fake Title',
        hole = 0.5)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
fig = px.pie(mar_amount,
      names = 'MaritalStatus',
      values = 'AmountPaid',
      width = 400,
      height = 400,
      title = 'Fake Title',
      hole = 0.5,
      color_discrete_sequence = px.colors.sequential.GnBu)
fig.update_traces(textposition='inside', textinfo='percent+label')

fig.show()
```

```python
fig = px.pie(mar_amount,
      names = 'MaritalStatus',
      values = 'AmountPaid',
      width = 400,
      height = 400,
      title = 'Fake Title',
      hole = 0.5,
      color_discrete_map = {'Single':'lightcyn',
                                 'Married':'lightpink',
                                 'Divorced':'lightblue'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
