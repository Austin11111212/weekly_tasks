```python
class Account:
    def __init__(self, account_number, account_holder):
        self.account_number = account_number
        self.account_holder = account_holder
        self.my_acc = {}
        self.amount = 0

    def introduce_self(self):
      print(f'{self.account_number}, account_name {self.account_holder}')

    def acc_details(self):
      self.my_acc['acc_num'] = self.account_number
      self.my_acc['acc_name'] = self.account_holder
      self.my_acc['acc_bal'] = self.amount

      print(f'account details: {self.my_acc}')


    def deposit(self, amount):
        self.amount += float(amount)
        self.my_acc['acc_bal'] = self.amount
        print(f'Deposited {self.amount}. New balance: {amount}')

    def withdraw(self, amount):
        if amount <= self.amount:
            self.amount -= float(amount)
            self.my_acc['acc_bal'] = self.amount
            print(f'Withdrew {amount}, New balance: {self.amount}')
        else:
            print('Insufficient funds.')

    def check_balance(self):
        return self.amount
```

```python
augustus = Account('0023345674', 'augustus')
```

```python
augustus.acc_details()
```

```python
augustus.deposit(50000)
```

```python
augustus.check_balance()
```

```python
augustus.withdraw(30000)
```
