```python
import requests
from IPython.display import display, Image

# Replace 'your_api_key_here' with the actual API key you received
api_key = 'ssPghEJ4IzZDN0Pp0wHv4Oxh12768KduhTpkeyiH' # Make sure this is your valid API key

# Define the APOD API endpoint URL
apod_url = 'https://api.nasa.gov/planetary/apod'

# Send a GET request to the API
response = requests.get(apod_url, params={'api_key': api_key})

# Check if the request was successful
if response.status_code == 200:
    apod_data = response.json()

    # Extract the image URL
    image_url = apod_data.get('url')

    # Display the image
    if image_url:
        display(Image(url=image_url))
    else:
        print("No image URL found in the response.")

else:
    print("Failed to retrieve data:", response.status_code) # This will help debug API issues
```
