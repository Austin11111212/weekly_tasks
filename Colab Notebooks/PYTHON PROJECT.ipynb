```python
questions = [
    {
        "question": "What is the capital of France?",
        "choices": ["A. Berlin", "B. Madrid", "C. Paris", "D. Rome"],
        "answer": "C"
    },
    {
        "question": "What is 2 + 2?",
        "choices": ["A. 3", "B. 4", "C. 5", "D. 6"],
        "answer": "B"
    }
    # Add more questions as needed
]
```

```python
def ask_question(question_data):
    print(question_data["question"])
    for choice in question_data["choices"]:
        print(choice)

    user_answer = input("Please enter your answer (A, B, C, D): ").strip().upper()

    if user_answer == question_data["answer"]:
        print("Correct!")
        return True
    else:
        print("Incorrect. The correct answer was", question_data["answer"])
        return False
```

```python
import random

def run_quiz():
    score = 0
    random.shuffle(questions)  # Randomize the order of questions

    for question in questions:
        if ask_question(question):
            score += 1

    print(f"Your final score is {score}/{len(questions)}")

# Start the quiz
run_quiz()
```

```python
import json

def save_score(score, file_name="high_scores.json"):
    try:
        with open(file_name, "r") as file:
            high_scores = json.load(file)
    except FileNotFoundError:
        high_scores = []

    high_scores.append(score)
    high_scores = sorted(high_scores, reverse=True)[:5]  # Keep top 5 scores

    with open(file_name, "w") as file:
        json.dump(high_scores, file)

def display_high_scores(file_name="high_scores.json"):
    try:
        with open(file_name, "r") as file:
            high_scores = json.load(file)
    except FileNotFoundError:
        high_scores = []

    print("High Scores:")
    for i, score in enumerate(high_scores):
        print(f"{i+1}. {score}")

def run_quiz():
    score = 0
    random.shuffle(questions)

    for question in questions:
        if ask_question_with_timer(question):
            score += 1

    print(f"Your final score is {score}/{len(questions)}")
    save_score(score)
    display_high_scores()

# Start the quiz
run_quiz()
```

```python
import tkinter as tk
from tkinter import messagebox

class QuizApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Interactive Python Quiz Game")

        self.question_index = 0
        self.score = 0

        self.question_label = tk.Label(root, text="", wraplength=400)
        self.question_label.pack(pady=20)

        self.choice_vars = [tk.StringVar(value="") for _ in range(4)]
        self.choice_buttons = [tk.Radiobutton(root, textvariable=self.choice_vars[i], value=chr(65+i)) for i in range(4)]
        for button in self.choice_buttons:
            button.pack(pady=5)

        self.submit_button = tk.Button(root, text="Submit", command=self.check_answer)
        self.submit_button.pack(pady=20)

        self.load_question()

    def load_question(self):
        if self.question_index < len(questions):
            q = questions[self.question_index]
            self.question_label.config(text=q["question"])
            for i, choice in enumerate(q["choices"]):
                self.choice_vars[i].set(choice)
        else:
            messagebox.showinfo("Quiz Complete", f"Your final score is {self.score}/{len(questions)}")
            self.root.quit()

    def check_answer(self):
        selected_answer = next(var.get() for var in self.choice_vars if var.get().startswith(tk.Radiobutton().cget('value')))
        correct_answer = questions[self.question_index]["answer"]
        if selected_answer == correct_answer:
            self.score += 1
        self.question_index += 1
        self.load_question()

if __name__ == "__main__":
  root = tk.Tk()
  app = QuizApp(root)
  root.mainloop()
```
