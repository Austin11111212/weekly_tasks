```python
# Create a Series with prices of the products
prices = [2, 1, 3, 4]
```

```python
prices
```

```python
# Convert the list into a pandas Series
price_series = pd.Series([2,1,3,4])
```

```python
price_series
```

```python
price_series[2]
```

```python
price_series[2:4]
```

```python
# Add the two Series together
total_series = product_series + ": $" + price_series.astype(str)
```

```python
total_series
```
