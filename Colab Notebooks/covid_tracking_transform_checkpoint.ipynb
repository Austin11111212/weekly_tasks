```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/covid_tracking.csv')
```

```python
df.head(5)
```

```python
# Rename the 'dataRep' column to 'Date'
df.rename(columns={'dataRep': 'date'}, inplace=True)
```

Duplicate the Date Column

```python
# Duplicate the 'Date' column
df['date_Copy'] = df['date']
```

 Split the Date Column into Day, Month, and Year


```python
# Replace hyphens with forward slashes in 'date' column
df['date'] = df['date'].str.replace('-', '/')

# Split the 'Date' column into 'Day', 'Month', and 'Year'
df[['Year', 'Month', 'Day']] = df['date'].str.split('/', expand=True)

# Convert these new columns to numeric types
df['Day'] = pd.to_numeric(df['Day'], errors='coerce')
df['Month'] = pd.to_numeric(df['Month'], errors='coerce')
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
```

Verify Field Types

```python
# Check the data types
print(df.dtypes)

# Verify that 'Day', 'Month', and 'Year' are integers
print(df[['Day', 'Month', 'Year']].head())
```

Replace Null Values with 0

```python
# Replace null values with 0
df.fillna(0, inplace=True)
```

 -->Rename Columns with Meaningful Names

```python
# Rename columns
df.rename(columns={
    'Date_Copy': 'Original_Date',
    'Day': 'Day_of_Month',
    'Month': 'Month_of_Year',
    'Year': 'Year'
}, inplace=True)
```

```python
# Save the transformed dataset to a new CSV file if needed
df.to_csv('transformed_covid_data.csv', index=False)
```
