````python
def most_common_char(string):
    # Initialize a dictionary to count the occurrences of each character
    char_count = {}

    # Count occurrences of each character in the string
    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    # Find the character with the maximum count
    max_count = 0
    most_common = None
    for char, count in char_count.items():
        if count > max_count:
            max_count = count
            most_common = char

    return most_common

# Example usage:
string = "hello world"
print(most_common_char(string))
````

Functions Challenge 9  Is a prime number?


Write a function called is_prime that takes an integer as an argument and checks if it is a prime number or not, returning a Boolean.

````python
def is_prime(number):
    if number <= 1:  # Numbers less than or equal to 1 are not prime
        return False
    elif number <= 3:  # 2 and 3 are prime
        return True
    elif number % 2 == 0 or number % 3 == 0:  # Even numbers greater than 2 and multiples of 3 are not prime
        return False
    else:
        # Check divisibility by numbers of the form 6k  1, where k is a non-negative integer
        # This is a common optimization for prime checking
        for i in range(5, int(number ** 0.5) + 1, 6):
            if number % i == 0 or number % (i + 2) == 0:
                return False
        return True

# Example usage:
print(is_prime(7))  # Output will be True
print(is_prime(12))  # Output will be False
````

Functions Challenge 11  Highest Common Factor


Write a function called hcf that takes 2 integer values and returns the highest common factor of the numbers.

````python
def hcf(num1, num2):
    # Find the smaller number between num1 and num2
    smaller = min(num1, num2)

    # Initialize HCF to 1
    hcf = 1

    # Iterate from 1 to the smaller number
    for i in range(1, smaller + 1):
        # Check if both num1 and num2 are divisible by i
        if num1 % i == 0 and num2 % i == 0:
            # Update HCF if i is a common factor
            hcf = i

    return hcf

# Example usage:
print(hcf(24, 36))  # Output will be 12
````

Functions Challenge 11  Valid Date


Write a function called is_valid_date that takes a date in string format DD/MM/YYYY and checks to see if it is a valid date and in the correct format.

Your function should take account of leap years.

````python
def is_valid_date(date):
    try:
        day, month, year = map(int, date.split('/'))

        # Check if year is a leap year
        if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
            leap_year = True
        else:
            leap_year = False

        # Dictionary containing the number of days in each month
        days_in_month = {
            1: 31, 2: 29 if leap_year else 28, 3: 31, 4: 30,
            5: 31, 6: 30, 7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31
        }

        # Check if month and day are in valid range
        if month not in days_in_month or day < 1 or day > days_in_month[month]:
            return False

        return True

    except ValueError:
        # If the date cannot be split or converted to integers, it's not in the correct format
        return False

# Example usage:
print(is_valid_date("29/02/2020"))  # Output will be True
print(is_valid_date("31/04/2022"))  # Output will be False
````

````python
def is_valid_postcode(postcode):
    # Check if the length of the postcode is correct
    if len(postcode) != 7:
        return False

    # Check if the first two characters are letters
    if not postcode[:2].isalpha():
        return False

    # Check if the next 1 or 2 characters are digits
    if not postcode[2:4].isdigit() or len(postcode[2:4]) > 2:
        return False

    # Check if there's a space at the fourth position
    if postcode[4] != ' ':
        return False

    # Check if the fifth character is a digit
    if not postcode[5].isdigit():
        return False

    # Check if the last two characters are letters
    if not postcode[-2:].isalpha():
        return False

    return True

# Example usage:
print(is_valid_postcode("AB12 3CD"))  # Output will be True
print(is_valid_postcode("A123 4BC"))  # Output will be False
````

Functions Challenge 14  Check Valid Email Address


Write a function called is_valid_email  that takes an email address as an argument and returns True/False depending on whether it is a valid email address.

Check rules:

Must contain at least 1 character before the at symbol
Must contain an @ symbol
Must have at-least 1 character after the @ symbol and before the period(.)
Must contain at least 1 character after the last period(.).
Maximum 256 characters
Must start with a letter or a number

````python
def is_valid_email(email):
    # Check if the length of the email is within the valid range
    if len(email) > 256:
        return False

    # Split the email address into local part and domain part
    local_part, domain_part = email.split('@', 1)

    # Check if local part and domain part have at least 1 character
    if not local_part or not domain_part:
        return False

    # Check if local part starts with a letter or a number
    if not local_part[0].isalnum():
        return False

    # Check if domain part contains at least one character before and after the period
    if '.' not in domain_part or domain_part.startswith('.') or domain_part.endswith('.'):
        return False

    # Check if there's at least one character before and after the period in domain part
    if not any(part.isalpha() for part in domain_part.split('.')):
        return False

    return True

# Example usage:
print(is_valid_email("user@example.com"))  # Output will be True
print(is_valid_email("user@.com"))         # Output will be False
````

Functions Challenge 15  Caesar Cipher Encryptor / Decryptor


Create a function called encrypt that takes some text(String) and a Shift(Integer)and then encrypts the text using the Caesar Cipher algorithm, returning the encrypted text.

Create a second function to decrypt an encrypted string, using the same input parameters and returning the decrypted text.

````python
def encrypt(text, shift):
    encrypted_text = ""
    for char in text:
        # Encrypt uppercase letters
        if char.isupper():
            encrypted_text += chr((ord(char) + shift - 65) % 26 + 65)
        # Encrypt lowercase letters
        elif char.islower():
            encrypted_text += chr((ord(char) + shift - 97) % 26 + 97)
        # Leave non-alphabetic characters unchanged
        else:
            encrypted_text += char
    return encrypted_text

def decrypt(text, shift):
    decrypted_text = ""
    for char in text:
        # Decrypt uppercase letters
        if char.isupper():
            decrypted_text += chr((ord(char) - shift - 65) % 26 + 65)
        # Decrypt lowercase letters
        elif char.islower():
            decrypted_text += chr((ord(char) - shift - 97) % 26 + 97)
        # Leave non-alphabetic characters unchanged
        else:
            decrypted_text += char
    return decrypted_text

# Example usage:
original_text = "Hello, World!"
shift = 3
encrypted_text = encrypt(original_text, shift)
print("Encrypted text:", encrypted_text)  # Output will be: Khoor, Zruog!
decrypted_text = decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)  # Output will be: Hello, World!
````

Functions Challenge 16  Find all palindromes


Create a function called get_palindrones that is given a string and returns a list of all the palindromes in the string.

````python
def is_palindrome(s):
    return s == s[::-1]

def get_palindromes(string):
    palindromes = []
    words = string.split()
    for word in words:
        if is_palindrome(word):
            palindromes.append(word)
    return palindromes

# Example usage:
text = "level radar hello kayak world civic"
print(get_palindromes(text))
# Output will be: ['level', 'radar', 'kayak', 'civic']
````

Functions challenge 17  Sort Competitors List


Create a function that takes a 2D list of competitor names and league (e.g    Bob Jones, Bronze) as well as an optional parameter Reverse.

The function should return the list sorted by league. (the league senority is Bronze, Silver, Gold, Ruby, Diamond), reversed if the optional reversed parameter is set to true.



````python
def sort_competitors_list(competitors_list, reverse=False):
    # Define the league hierarchy
    league_order = {"Bronze": 1, "Silver": 2, "Gold": 3, "Ruby": 4, "Diamond": 5}

    # Sort the competitors list based on the league hierarchy
    sorted_list = sorted(competitors_list, key=lambda x: league_order.get(x[1], 0))

    # Reverse the sorted list if the 'reverse' parameter is True
    if reverse:
        sorted_list.reverse()

    return sorted_list

# Example usage:
competitors = [
    ["Bob Jones", "Silver"],
    ["Alice Smith", "Gold"],
    ["John Doe", "Bronze"],
    ["Jane Doe", "Ruby"],
    ["Mike Johnson", "Diamond"]
]

sorted_competitors = sort_competitors_list(competitors)
print("Sorted by league (ascending):", sorted_competitors)

reversed_sorted_competitors = sort_competitors_list(competitors, reverse=True)
print("Sorted by league (descending):", reversed_sorted_competitors)
````
