```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Sample dataset of transactions
data = [
    ['milk', 'bread', 'butter','yam'],
    ['bread', 'butter'],
    ['milk', 'bread','egg','fish'],
    ['milk', 'butter','egg'],
    ['bread'],
    ['milk', 'bread', 'butter'],
    ['bread'],
    ['milk', 'bread'],
    ['bread', 'butter'],
    ['milk', 'bread', 'yam']
]
```

```python
data
```

```python
# Convert the dataset to a DataFrame
df = pd.DataFrame(data)
```

```python
df
```

```python
# Initialize the TransactionEncoder
te = TransactionEncoder()

# Fit and transform the data
te_ary = te.fit_transform(data)
```

```python
te_ary
```

```python
# Convert the transformed data to a DataFrame
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df
```

```python
frequent_itemsets=apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets.sort_values('support',ascending=False)
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2).sort_values('confidence',ascending=False)
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2).sort_values('lift',ascending=False)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('DataSetA.csv',on_bad_lines='skip')
```

```python
df.shape
```

```python
df.head(10)
```

```python
transactions = []
```

```python
for index, row in df.iterrows():
    transaction = [value for value in row if pd.notna(value)]
    transactions.append(transaction)
```

```python
transactions
```

```python
# Initialize the TransactionEncoder
te = TransactionEncoder()

# Fit and transform the data
te_ary = te.fit_transform(transactions)
```
