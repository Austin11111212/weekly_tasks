```python
import pandas as pd
import numpy as mp
import plotly.express as px
```

```python
df = pd.read_csv('/content/spotify-2023 2.csv', encoding = 'Latin')
```

```python
df.head()
```

```python
fig = px.pie(df,values = 'streams',names = 'released_month',hole = 0.25)
fig.update_traces(textposition='inside', textinfo='percent+label')
```

```python
fig.show()
```

```python
df.sort_values(by = 'artist_count', ascending = False, inplace = True)
sample_data = df.head(15)
```

```python
sample_data
```

```python
fig = px.bar(sample_data, y = 'artist(s)_name',x = 'artist_count', width = 500, height = 500, title = 'spotify data')
fig.show()
```

```python
df['streams'] = pd.to_numeric(df['streams'],errors = 'coerce')
```

```python
df.sort_values(by = 'streams', ascending = False, inplace = True)
```

```python
fig = px.scatter(df,x = 'streams',y = 'released_day',color = 'released_month')
fig.show()
```

```python
df['streams'].mode()
```

```python
df['released_month'].value_counts().reset_index()
```

```python
fig = px.scatter(df,x = 'streams',y = 'released_day',size = 'released_month',color = 'released_month')
fig.show()
```

```python
df = df.groupby('artist_count')['streams'].mean().reset_index().sort_values(by = 'artist_count', ascending = True).reset_index(drop=True)
df = df.head()
```

```python
fig= px.funnel(df,y= 'streams',x = 'artist_count')
fig.show()
```
