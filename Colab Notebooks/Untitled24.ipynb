```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({
    'category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'amt': [100, 200, 300, 150, 250, 350]
})

# Calculate the average and sum of 'amt' for each category
summary_df = df.groupby('category')['amt'].agg(['mean', 'sum']).reset_index()

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for average
ax.bar(summary_df['category'] - 0.2, summary_df['mean'], width=0.4, label='Average', color='blue')

# Bar plot for sum
ax.bar(summary_df['category'] + 0.2, summary_df['sum'], width=0.4, label='Sum', color='green')

ax.set_xlabel('Category')
ax.set_ylabel('Amount')
ax.set_title('Average and Sum of Amount by Category')
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({
    'category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'amt': [100, 200, 300, 150, 250, 350]
})

# Calculate the average and sum of 'amt' for each category
summary_df = df.groupby('category')['amt'].agg(['mean', 'sum']).reset_index()

# Convert 'category' column to numerical representation for plotting
# We'll map categories to numbers (A:0, B:1, C:2)
category_mapping = {'A': 0, 'B': 1, 'C': 2}
summary_df['category_num'] = summary_df['category'].map(category_mapping)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for average
ax.bar(summary_df['category_num'] - 0.2, summary_df['mean'], width=0.4, label='Average', color='blue')

# Bar plot for sum
ax.bar(summary_df['category_num'] + 0.2, summary_df['sum'], width=0.4, label='Sum', color='green')

# Set x-axis ticks and labels back to original categories
ax.set_xticks(summary_df['category_num'])
ax.set_xticklabels(summary_df['category'])

ax.set_xlabel('Category')
ax.set_ylabel('Amount')
ax.set_title('Average and Sum of Amount by Category')
ax.legend()

plt.show()
```
