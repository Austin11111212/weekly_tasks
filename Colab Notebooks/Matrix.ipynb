```python
import numpy as np

# Create a 2x3 matrix with random numerical data
matrix = np.random.rand(2, 3)
print("Matrix:")
print(matrix)

# Identify the data type of the matrix
data_type = matrix.dtype
print("Data type of the matrix:", data_type)

# Calculate mean, mode, and median
mean = np.mean(matrix)
mode = np.median(matrix)
median = np.median(matrix)
print("Mean of the matrix:", mean)
print("Mode of the matrix:", mode)
print("Median of the matrix:", median)

# Basic matrix operations
# Addition
matrix_addition = matrix + matrix
print("Matrix Addition:")
print(matrix_addition)

# Subtraction
matrix_subtraction = matrix - matrix
print("Matrix Subtraction:")
print(matrix_subtraction)

# Transpose
matrix_transpose = np.transpose(matrix)
print("Matrix Transpose:")
print(matrix_transpose)

# Scalar Multiplication
scalar = 2
matrix_scalar_multiplication = scalar * matrix
print("Scalar Multiplication:")
print(matrix_scalar_multiplication)
```
