Write a python code to collect information from the user and print all the vowels in that info.



```python
def extract_vowels(text):
    vowels = "aeiouAEIOU"
    vowel_list = [char for char in text if char in vowels]
    return vowel_list

def main():
    user_info = input("Please enter some text: ")
    vowels_in_info = extract_vowels(user_info)

    if vowels_in_info:
        print('Vowels in the provided text are:', ' '.join(vowels_in_info))
    else:
        print('No vowels found in the provided text.')

if __name__ == "__main__":
    main()
```

Write a python code to extract all the common letters between 2 strings and store them in a list.

```python
def common_letters(string1, string2):
    common = []
    for char in string1:
        if char in string2 and char not in common:
            common.append(char)
    return common

def main():
    string1 = input("Enter the first string: ")
    string2 = input("Enter the second string: ")

    common_letters_list = common_letters(string1, string2)

    if common_letters_list:
        print("Common letters between the two strings:", ' '.join(common_letters_list))
    else:
        print("No common letters found between the two strings.")

if __name__ == "__main__":
    main()
```

Write a python code to generate the sum of all the numbers that are divisible by 2 and 5 between 1 and 50.

```python
def sum_divisible_by_2_and_5(start, end):
    total_sum = 0
    for num in range(start, end + 1):
        if num % 2 == 0 and num % 5 == 0:
            total_sum += num
    return total_sum

def main():
    start = 1
    end = 50
    result = sum_divisible_by_2_and_5(start, end)
    print("Sum of numbers divisible by 2 and 5 between 1 and 50:", result)

if __name__ == "__main__":
    main()
```

 Write a Python program that calculates the Average Grade Point (AGP)of a student based on the scores entered by the user.
The program should:
i) prompt the user to enter the number of subjects
 ii) prompt the user to enter the scores for each subject,
iii) calculate the average grade using the formula: Average Grade = Total Score / Number of Subjects and
iv) display the average grade to the user.


```python
def calculate_average_grade_point():
    # Prompt the user to enter the number of subjects
    num_subjects = int(input("Enter the number of subjects: "))

    # Initialize total score to zero
    total_score = 0

    # Prompt the user to enter the scores for each subject
    for i in range(1, num_subjects + 1):
        score = float(input(f"Enter the score for subject {i}: "))
        total_score += score

    # Calculate the average grade
    average_grade = total_score / num_subjects

    return average_grade

def main():v
    average_grade = calculate_average_grade_point()
    print("Average Grade Point (AGP):", average_grade)

if __name__ == "__main__":
    main()
```
