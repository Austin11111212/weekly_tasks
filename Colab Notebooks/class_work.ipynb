```python
import pandas as pd

# Load the dataset
data = pd.read_csv('tourism_dataset.csv')

# Identify columns with missing values
missing_values = data.isnull().sum()
print('Columns with missing values:')
print(missing_values[missing_values > 0])
```

```python
# Handle missing values (for example, fill with mean)
data.fillna(data.mean(), inplace=True)
```

```python
# Check for duplicates
duplicates = data.duplicated().sum()
print("Number of duplicate rows:", duplicates)

# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# Remove irrelevant columns
irrelevant_columns = ["column1", "column2"]  # Add irrelevant column names
data.drop(columns=irrelevant_columns, inplace=True)
```

```python
# Summary statistics
summary_stats = data.describe()
print(summary_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
correlation_matrix = data.corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Most frequent countries
top_countries = data['country'].value_counts().head(3)
print("Top 3 most frequent countries:")
print(top_countries)
```

```python
# Most common purpose of visit by age_group
common_purpose_by_age = data.groupby('age_group')['purpose'].agg(lambda x: x.value_counts().index[0])
print("Most common purpose of visit by age_group:")
print(common_purpose_by_age)
```

```python
# Data distribution of 'total_cost' column by Country
plt.figure(figsize=(10, 6))
sns.boxplot(x='country', y='total_cost', data=data)
plt.title("Total Cost Distribution by Country")
plt.show()

# Data distribution of 'total_cost' column by age_group
plt.figure(figsize=(10, 6))
sns.boxplot(x='age_group', y='total_cost', data=data)
plt.title("Total Cost Distribution by Age Group")
plt.show()

# Data distribution of 'total_cost' column by purpose
plt.figure(figsize=(10, 6))
sns.boxplot(x='purpose', y='total_cost', data=data)
plt.title("Total Cost Distribution by Purpose")
plt.xticks(rotation=45)
plt.show()
```
