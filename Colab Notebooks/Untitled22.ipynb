Import Required Libraries

```python
import pandas as pd
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

Define the Toy Dataset

```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
               ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
               ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
               ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
               ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

Convert the Dataset into a Format Suitable for Apriori

```python
te = TransactionEncoder()
te_ary = te.fit(toy_dataset).transform(toy_dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

Run the Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
```

Generate Association Rules

```python
association_rules(frequent_itemsets, metric="lift", min_threshold=0.2).sort_values('confidence',ascending=False)
```

Interpret Results
Frequent Itemsets: These are groups of items that appear frequently together.
Association Rules: These rules tell us how the presence of one item can imply the presence of another.

Preprocessing:

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

# Assuming each row in 'df' represents a transaction
df.values.tolist()

# Convert all items in transactions to strings
transactions = [[str(item) for item in transaction] for transaction in transactions]
```

```python
transactions
```

```python
# Transaction Encoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
te
```

```python
te_ary
```

```python
df_encoded
```

Run the Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Find frequent itemsets with a minimum support of 0.2 (adjust as needed)
frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)
frequent_itemsets.sort_values('support',ascending=False)
```

```python
frequent_itemsets
```

```python
# Generate association rules with a minimum confidence of 0.3(adjust as needed)
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.3)
```
