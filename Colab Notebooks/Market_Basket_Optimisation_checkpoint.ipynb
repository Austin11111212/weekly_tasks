Step 1: Apriori Algorithm on Toy Dataset

 Load the Toy Dataset


```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
```

```python
# Toy dataset
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
               ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
               ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
               ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
               ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
toy_dataset
```

```python
# Convert the dataset to a DataFrame
df = pd.DataFrame(toy_dataset)
```

```python
df
```

```python
# Convert dataset into a one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit_transform(toy_dataset)
```

```python
te_ary
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df
```

```python
frequent_itemsets=apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets.sort_values('support',ascending=False)
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2).sort_values('confidence',ascending=False)
```

 Interpret the Results


Frequent Itemsets: These are combinations of items that appear frequently together in transactions.
Association Rules: These show relationships between items, such as which items tend to be bought together. Each rule has a confidence value, which indicates the likelihood of the consequent item being bought when the antecedent item is bought.
For instance, if you find a rule like {Skirt, Scarf} -> {Pants} with high confidence, it suggests that customers who buy Skirts and Scarves are also likely to buy Pants.

Run Apriori Algorithm on Checkpoint Dataset

```python
transactions = []
```

```python
for index, row in df.iterrows():
    transaction = [value for value in row if pd.notna(value)]
    transactions.append(transaction)
```

```python
transactions
```

```python
# Initialize the TransactionEncoder
te = TransactionEncoder()

# Fit and transform the data
te_ary = te.fit_transform(transactions)
```

```python
te_ary
```

```python
# Convert the transformed data to a DataFrame
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df.head(10)
```

```python
df.shape
```

```python
frequent_itemsets=apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets.sort_values('support',ascending=False)
```

```python
asso = association_rules(frequent_itemsets,metric='confidence',min_threshold=0.7
                  ).sort_values('confidence',ascending=False)

asso
```
