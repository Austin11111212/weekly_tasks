```python
# Create a dictionary to store items and their corresponding prices
item_prices = {
    "apple": 1.00,
    "banana": 0.50,
    "bread": 2.00,
    "milk": 1.50,
    "eggs": 2.50
}

# Create a list named 'shopping_list' to store the items
shopping_list = []

# Use a while loop to create a menu of options for the user
while True:
    print("\nMenu:")
    print("1. Add item")
    print("2. Remove item")
    print("3. View items")
    print("4. Exit")

    # Prompt the user to make a selection from the menu
    choice = input("Enter your choice: ")

    # Use an if-elif-else block to determine the user's selection
    if choice == '1':  # If the user selects 'add'
        if len(shopping_list) < 5:  # Limit the number of items that can be added
            item = input("Enter the item to add: ").lower()
            if item in item_prices:
                shopping_list.append(item)
                print(f"'{item}' has been added to the shopping list.")
            else:
                print("Item not found in the price list.")
        else:
            print("Maximum limit reached. You cannot add more items.")
    elif choice == '2':  # If the user selects 'remove'
        item = input("Enter the item to remove: ").lower()
        if item in shopping_list:
            shopping_list.remove(item)
            print(f"'{item}' has been removed from the shopping list.")
        else:
            print(f"'{item}' is not in the shopping list.")
    elif choice == '3':  # If the user selects 'view'
        print("Shopping List:")
        total_price = 0
        for index, item in enumerate(shopping_list, start=1):
            price = item_prices.get(item, 0)
            total_price += price
            print(f"{index}. {item.capitalize()} - ${price:.2f}")
        print(f"Total Price: ${total_price:.2f}")
    elif choice == '4':  # If the user selects 'exit'
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```

```python
# Define basic mathematical functions
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Division by zero!"
    return x / y

# Dictionary mapping operation symbols to functions
operations = {
    '+': add,
    '-': subtract,
    '*': multiply,
    '/': divide
}

# Calculator function
def calculator():
    num1 = float(input("Enter the first number: "))

    print("Available operations:")
    for symbol in operations:
        print(symbol)
    while True:
        operation_symbol = input("Choose an operation symbol: ")

        num2 = float(input("Enter the second number: "))

        calculation_function = operations[operation_symbol]
        answer = calculation_function(num1, num2)

        print(f"{num1} {operation_symbol} {num2} = {answer}")

        choice = input("Do you want to continue with this result? (yes/no): ")
        if choice.lower() == 'yes':
            num1 = answer
        else:
            break

            calculator()  # Start a new calculation

# Start the calculator
calculator()
```
